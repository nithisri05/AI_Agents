# CrewAI Blog Generator Agent

This notebook demonstrates how to build a **single-agent system using CrewAI** that generates professional **LinkedIn blog posts**.



## Objective
We will:
- Create an AI Agent (Blog Writer)
- Define a Task (Generate LinkedIn Blog)
- Use CrewAI to execute the workflow
- Securely load API credentials using Colab Secrets



## Workflow

User Input → Task → Agent → LLM → Blog Output



## Tech Stack
- CrewAI
- LLM (via API)
- Google Colab Secrets

## Load Required Libraries & Secrets

We import:
- CrewAI components
- OS utilities
- Colab secrets to securely fetch API credentials


```python
from crewai import LLM, Agent, Task, Crew
import os
from google.colab import userdata

# Load secrets
api_key = userdata.get('OPENAI_API_KEY')
base_url = userdata.get('OPENAI_BASE_URL')
```

## Initialize the Language Model (LLM)

This defines:
- Which model to use
- Creativity level (temperature)
- API endpoint

The LLM is the **brain of the agent**

```python
llm = LLM(
    model="gpt-4.1-nano",
    temperature=0.7,
    base_url=base_url,
    api_key=api_key
)
```

## Define the Agent (Blog Writer)

An Agent represents:
- Role → Who it is
- Goal → What it aims to achieve
- Backstory → Behavior/personality

Here, we create a **LinkedIn Blog Writer Agent**

```python
blog_writer = Agent(
    role="AI Blog Writer",
    goal="Create high-quality, professional LinkedIn blog posts on AI and tech topics",
    backstory="An experienced technical writer who creates engaging, insightful, and structured blog posts for professionals.",
    tools=[],
    llm=llm,
)
```

## User Input

The user provides a topic for the blog.

Example:
- Agentic AI
- RAG vs Fine-tuning
- Future of AI Agents

```python
topic = input("Enter topic for LinkedIn blog: ")
```

## Define the Task

A Task specifies:
- What the agent should do
- Output format
- Instructions

We instruct the agent to generate a **LinkedIn-ready blog post**

```python
blog_task = Task(
    description=f"""
    Write a professional LinkedIn blog post on the topic: '{topic}'.

    The blog should include:
    - Strong hook (first 2 lines engaging)
    - Clear explanation of the concept
    - Real-world relevance
    - Structured sections
    - Bullet points where needed
    - Ending with a thoughtful conclusion or question

    Keep it professional, insightful, and easy to read.
    """,

    expected_output="""
    A well-structured LinkedIn blog post with:
    - Hook
    - Body (sections or paragraphs)
    - Bullet points (if needed)
    - Conclusion
    """,

    agent=blog_writer
)
```

## Run the Crew

Crew is the orchestrator:
- Assigns tasks to agents
- Executes workflow
- Returns output

Here we run the system and generate the blog

```python
crew = Crew(tasks=[blog_task])

result = crew.kickoff()

print(result)
```
